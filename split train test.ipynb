```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

df = pd.read_csv('data/quora_duplicate_questions.tsv', 
                 sep ='\t',index_col = 0  ).dropna() #had to drop na, 1 in queston1 and 2 in question2



X = df[['question1','question2']]
y = df['is_duplicate']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
training = X_train.copy()
training['is_duplicate'] = y_train.copy()

testing = X_test.copy()
testing['is_duplicate'] = y_test.copy()
```

```python
training.to_csv('data/training.csv')
testing.to_csv('data/testing.csv')
```
